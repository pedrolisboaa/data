```python
import json
import pandas as pd

caminho = 'anac/V_OCORRENCIA_AMPLA.json'

# Abrindo corretamente com BOM
with open(caminho, 'r', encoding='utf-8-sig') as f:
    dados = json.load(f)

# Se for dict, verifica as chaves
if isinstance(dados, dict):
    for chave in dados:
        print(f'Chave: {chave} | Tipo: {type(dados[chave])}')
        if isinstance(dados[chave], list):
            df = pd.DataFrame(dados[chave])
            break
elif isinstance(dados, list):
    df = pd.DataFrame(dados)
else:
    raise ValueError("Formato JSON inesperado.")

# Visualiza resultado
df.head()
```
